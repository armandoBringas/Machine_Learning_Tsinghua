```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
data_dir = Path('../data/processed/train/038441c925bb/')

# Iterate over each .npy file in the directory
for file_path in data_dir.glob('*.npy'):
    # Load the series data from the .npy file
    series_data = np.load(file_path)

    # Create a plot
    plt.figure(figsize=(20, 2))  # Adjust the size as needed
    plt.plot(series_data, label=file_path.stem)  # file_path.stem gives the file name without extension

    # Customize the plot
    plt.title(f'Time Series Plot for {file_path.stem}')
    plt.xlabel('Time')
    plt.ylabel('Value')
    plt.legend()
    plt.show()
```
